```python

# 주차장명이 없는 CASE
row, col = park.shape
for i in range(row):
    # 주차장명이 float(nan) 인 경우 해당 row를 삭제
    if type(park['주차장명'][i]) == float:
        park.drop([i], inplace=True)
```

```python

# 개방시작, 개방종료 시간 양식 통일화
# ex1) 0:00 -> 00:00
# ex2) 미개방 -> nan
for cols in park.keys():
    if '개방' in cols:
        for i in park.index:
            try:
                len(park[cols][i])
            except:
                1
            else:
                if len(park[cols][i]) == 4 and ':00' in park[cols][i]:
                    park[cols][i] = '0' + park[cols][i]
                elif '24:00' in park[cols][i]:
                    park[cols][i] = '24:00'
            print("num!", i)
```
